```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
covid_data = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
# Explore data
global_cases = covid_data.iloc[:, 4:].sum()
global_cases.plot(figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Global COVID-19 Cases')
plt.show()
```

```python
# Analyze impact on different countries and regions
countries = ['US', 'India', 'Brazil', 'Russia', 'South Africa']
countries_data = covid_data[covid_data['Country/Region'].isin(countries)]
countries_cases = countries_data.iloc[:, 4:].sum()
countries_cases.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases by Country')
plt.show()
```

```python
sns.scatterplot(x='Confirmed', y='Deaths', hue='Country/Region', data=countries_data)
plt.xlabel('Number of Confirmed Cases')
plt.ylabel('Number of Deaths')
plt.title('COVID-19 Deaths vs. Confirmed Cases by Country')
plt.show()

```

```python
# Analyze impact on different regions
region_data = covid_data.groupby('WHO Region').sum().iloc[:, 3:]
region_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('WHO Region')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases by WHO Region')
plt.show()
```

```python
# Analyze daily new cases
daily_cases = covid_data.iloc[:, 4:].diff(axis=1)
daily_cases_mean = daily_cases.mean()
daily_cases_mean.plot(figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Number of Daily Cases')
plt.title('Global Daily New COVID-19 Cases')
plt.show()
```

```python
sns.boxplot(data=daily_cases.T, whis=1.5)
plt.xlabel('Day')
plt.ylabel('Number of Daily Cases')
plt.title('Distribution of Daily New COVID-19 Cases')
plt.show()
```
